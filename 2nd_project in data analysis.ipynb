```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Water Samples were collected from 13 locations along River Nile and tested for mercury.')

# Load the Excel file into a pandas DataFrame
df = pd.read_excel(r'D:\D.T\geology\G408\2nd_assignment.xlsx')

print(df.columns)

mercury_level_data = df['Mercury level']

# 2. Calculate Mean, Median, Mode for Iron oxide and Aluminum oxide element concentrations
mercury_mean = np.mean(mercury_level_data)
mercury_median = np.median(mercury_level_data)
mercury_mode = df['Mercury level'].mode().values[0]  # Mode might have multiple values, so we select the first one

print("mercury_mean =", mercury_mean)
print("mercury_median =", mercury_median)
print("mercury_mode =", mercury_mode)

# Compute the first quartile and third quartile
quartiles = df.quantile([0.25, 0.75])

# Compute the variance
variance = df.var()

# Compute the standard deviation
std_deviation = df.std()

# Compute the range
data_range = df.max() - df.min()

# Print the results
print("Quartiles:")
print(quartiles)
print("\nVariance:")
print(variance)
print("\nStandard Deviation:")
print(std_deviation)
print("\nRange:")
print(data_range)

# Compute the first quartile (25th percentile) and the third quartile (75th percentile)
quartiles = df.quantile([0.25, 0.75])

#Plot the quartiles
#quartiles.T.plot(kind='bar', figsize=(6, 3))
#plt.title('First Quartile and Third Quartile')
# remove the # from the brevious lines and you will det a different form of Q visualization >> Histogram.
df.boxplot()
plt.title('Boxplot of Quartiles')
plt.xlabel('')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(['25th Percentile', '75th Percentile'])
plt.grid(True)
plt.show()
```
